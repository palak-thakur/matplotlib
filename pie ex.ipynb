```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ramchandra = [4, 5, 6, 7]
total = sum(ramchandra)
names = ['krishan' , 'murari' , 'ghanshyam', 'murli']
prop = [15, 22, 19, 16]
names.append('rest')
prop.append(100 - sum(prop))
print(names, prop)
```

```python
plt.pie(prop,                       # data must be 1D array like
        labels=names,                  # labels should be list
        autopct=lambda pc: f'{pc * total/100: .2f}')  #using f string
plt.show()
```

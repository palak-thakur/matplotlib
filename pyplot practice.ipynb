```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
name= np.array(["Naruto", "Armin", "Light", "Mob"])
en_mark= np.array([34, 70, 90, 63])
sc_mark= np.array([25, 89, 95, 70])
mt_mark= np.array([30, 90, 90, 75])
```

```python
#creating a simple bar plot of names v/s en_mark
clr= ["r", "b", "g", "m"]
marks= plt.bar(x= name, height= en_mark, color= clr, width=0.4)
plt.grid(visible= True)
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Names v/s english marks")
plt.show()
```

```python
#adjusting alignment
clr= ["c", "y", "g", "m"]
marks= plt.bar(x= name, height= en_mark, color= clr, width=-0.4, align= "edge") #for +ve width it will align on right side
plt.grid(visible= True)
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Names v/s english marks")
plt.show()
```

```python
#stacked bar plot
plt.bar(x= name, height= en_mark, width= 0.4, label="english marks")
plt.bar(x= name, height= sc_mark, width=0.4,bottom= en_mark, label= "science marks")
plt.legend()
plt.grid(visible= True)
plt.show()
```

```python
#grouped bar chart
plt.bar(x= name, height= en_mark, width= -0.4, align= "edge", label="english marks")
plt.bar(x= name, height= sc_mark, width=0.4, align="edge", label= "science marks")
plt.grid(visible= True)
plt.legend()
plt.show()
```

```python
#grouped chart for three values
xmark= np.arange(name.size)
wd= 1/4
plt.figure(figsize= (10,5))
plt.bar(x= xmark+ wd, height= en_mark, width= wd, label="english marks")
plt.bar(x= xmark, height= sc_mark, width=wd, label= "science marks")
plt.bar(x= xmark-wd, height= sc_mark, width=wd, label= "science marks")
plt.xticks(xmark, name, rotation= 30)
plt.xlabel("Names of students")
plt.ylabel("Marks")
plt.title("Names v/s Makrs")
plt.grid(visible= True)
plt.legend()
plt.show()
```

```python
#line and marker plot
plt.plot(name,
        en_mark,
         color='r',
        ls="-",
        lw=3,
        marker= "p",
         ms= 10,
         mfc= "g",
         label= "English"
        )
plt.legend()
plt.show()
```

```python
#line and marker plot with science and math marks
plt.plot(name,
        sc_mark,
        color='g',
        ls="-",
        lw=3,
        marker="s",
        ms= 10,
        mfc='b',
        label= "Science")
plt.plot(name,
        mt_mark,
        color="b",
        ls="--",
        lw=3,
        marker= "p",
        ms=10,
        mfc="g",
        label="Maths")
plt.legend()
plt.show()
```

```python
#plot without line
plt.plot(name,
        sc_mark,
        color='g',
        ls='None',
        marker="s",
        ms= 10,
        mfc='b',
        label= "Science")
plt.plot(name,
        mt_mark,
        color="b",
        ls='None',
        marker= "p",
        ms=10,
        mfc="g",
        label="Maths")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
#scatter plot
plt.scatter(name, en_mark, s= 10, c= "b", marker= "o")
plt.xlabel("names")
plt.ylabel("marks")
plt.title("names v/s marks")
plt.show()
```

```python
#scatter plot for two subjects
plt.figure(figsize= (10,5))
plt.scatter(name, sc_mark, s= 40, c= "b", marker= "o")
plt.scatter(name, mt_mark, s= 40, c= "y", marker= "p")
plt.xlabel("names")
plt.ylabel("marks")
plt.title("names v/s marks")
plt.show()
```

```python
#histogram
arr= np.array([2, 5, 7, 8, 9, 11, 12, 13, 15, 16, 17, 20])
arr
```

```python
plt.hist(arr, bins=5, color= 'y') #by default bin =10
plt.xlabel("value of array")
plt.ylabel("frequency")
plt.title("data distribution of array")
plt.show()
# (end-start)/bins is size of blocks, start+ size of block and so on and number of values in each block will be on y axis
```

```python
#already defined bin size
bn=[2, 8, 15, 20]
plt.hist(arr, bins=bn, color='g')
plt.xlabel("value of array")
plt.ylabel("frequency")
plt.title("data distribution of array")
plt.show()
```

```python
#pie chart
plt.pie(en_mark,
        explode= [0.1, 0, 0, 0],
       labels= name,
       autopct= lambda pc: round(pc*np.sum(en_mark)/100)) #pc is percentage of marks
plt.legend()
plt.title("name and marks")
plt.show()
```

```python
#box plot
ls1= [2, 4, 6, 8, 100]
ls2= [3, 5, 7, 9, 39]
ls3= [ls1, ls2]
plt.boxplot(ls3)
plt.show()
```

```python
#annotation and text
plt.scatter(ls1, ls2, marker="p", c='b', s=20)
plt.annotate("fatherthest point", (100,39), xytext= (60,39), arrowprops= dict(arrowstyle= "->"))
plt.text(20, 20, "Scatter plot with anotation", fontsize=15)
plt.show()
```

```python
#bar plot with marks on labels
barin= plt.bar(x=name, height=en_mark, color= ["r", "g", "y", "b", "c"], width= 0.4)
plt.grid(visible=True)
plt.bar_label(barin)
plt.show()
```

```python
#sine curve
x= np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x))
plt.xlabel("Angle[rad]")
plt.ylabel("sine")
plt.axhline()
plt.axvline()
```

```python
#cos curve
x= np.linspace(-np.pi, np.pi,200)
plt.plot(x, np.cos(x))
plt.xlabel("Angle[rad]")
plt.ylabel("sine")
plt.axhline()
plt.axvline()
```

```python
#multiple plot in one plot
x= np.linspace(-np.pi, np.pi, 200)
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.subplot(2,2,3)
plt.plot(x, np.sin(x), ls="--")
plt.subplot(2,2,4)
plt.plot(x, np.cos(x), ls="--")
plt.show()
```

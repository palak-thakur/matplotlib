# Matplotlib object oriented API

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
names = np.array(["Amit", "Raman", "Geeta", "Neeta", "Jay"])
en_marks = np.array([73, 60, 67, 65, 75])  # marks in English
sc_marks = np.array([65, 70, 75, 63, 60])  # Marks in Science
mt_marks = np.array([60, 75, 62, 63, 70])  # Marks in Maths
```

```python
# Line and  marker plot
fig, ax = plt.subplots()
ax.plot(names, en_marks, ls="--", lw=3, marker="P", mfc="r", ms=14)
ax.set_xlabel("Names of students")
ax.set_ylabel("Marks in English")
ax.set_title("Students Vs marks in English")
ax.grid(True)
plt.show()
```

```python
# Line and  marker plot(Marker only)
fig1, ax1 = plt.subplots()
ax1.plot(names, en_marks, ls="None", marker="P", mfc="r", ms=14)
ax1.set_xlabel("Names of students")
ax1.set_ylabel("Marks in English")
ax1.set_title("Students Vs marks in English")
ax1.grid(True)
plt.show()
```

```python
# bar plot
fig2, ax2 = plt.subplots()
clr = ["green", "blue", "red", "indigo", "yellow"]
axob = ax2.bar(x=names, height=en_marks, width=0.4, color=clr)
ax2.set_xlabel("Names of students")
ax2.set_ylabel("Marks in English")
ax2.set_title("Students Vs marks in English")
ax2.grid(True)
ax2.bar_label(axob)
plt.show()
```

```python
# Horizontal bar plot
fig3, ax3 = plt.subplots()
ax3.barh(y=names, width=en_marks, height=0.4, color=["green", "blue", "red", "indigo", "yellow"])
ax3.set_xlabel("Marks of students")
ax3.set_ylabel("Names of students")
ax3.set_title("Students Vs marks in English")
plt.show()
```

```python
# Scatterplot
fig4, ax4 = plt.subplots()
ax4.scatter([2, 7, 8, 9], [3, 7, 8, 1], s=20, c="green")
plt.show()
```

```python
# Pie chart
expl = (0, 0.1, 0, 0, 0)
col = ["green", "blue", "red", "indigo","yellow"]
fig5, ax5 = plt.subplots()
ax5.pie(en_marks,
        explode=expl,
        labels=names,
        autopct=lambda x: f'{round((x* np.sum(en_marks))/100)}',
        shadow=True,
        colors= col,
        startangle=90)
ax5.axis('equal') 
ax5.legend()
plt.show()
```

```python
# Box plot
l1 = np.array([2, 4, 8, 3, 5, 10, 24])
l2 = np.array([3, 9, 15, 21, 27, 100, 25])
l3 = [l1, l2]
fig6, ax6 = plt.subplots()
ax6.boxplot(l3)
plt.show()
```

```python
# Stacked bar chart
width = 0.35 
fig, ax = plt.subplots()
ax.bar(names, en_marks, width, label='English')
ax.bar(names, sc_marks, width, bottom=en_marks, label='Science')
ax.set_ylabel('Scores')
ax3.set_xlabel("Names of students")
ax.set_title('Scores of students')
ax.legend()
plt.show()
```

```python
# Grouped bar chart
x = np.arange(names.size)  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, en_marks, width, label='English')
rects2 = ax.bar(x + width/2, sc_marks, width, label= "Science")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by Students')
ax.set_xticks(x, names)
ax.legend()
plt.show()
```

# Sine plot 

```python
rad_ang = np.linspace(-np.pi, np.pi, 200)
fig12, ax12 = plt.subplots()
ax12.plot(rad_ang, np.sin(rad_ang))
ax12.set_xlabel("Angle in radians")
ax12.set_ylabel("Sine(x)")
ax12.set_title("Plot of Sine curve")
ax12.grid(True)
ax12.axhline()
ax12.axvline()
plt.show()
```

# Subplots in one plot

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].plot(names, en_marks, ls="--", lw=3, marker="P", mfc="r", ms=14)
ax[1].plot(names, en_marks, ls="--", lw=3, marker="P", mfc="r", ms=14)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0, 0].plot(names, en_marks, ls="--", lw=3, marker="P", mfc="r", ms=14)
ax[0, 1].plot(names, en_marks, ls="--", lw=3, marker="P", mfc="r", ms=14)
ax[1, 0].bar(x=names, height=en_marks, width=0.4, color=clr)
ax[1, 1].bar(x=names, height=en_marks, width=0.4, color=clr)
plt.show()
```

# Annotation & text

```python
fig, ax = plt.subplots()
ax.scatter([2, 7, 8, 9], [3, 7, 8, 1], s=20, c="green")
ax.annotate("Important Point", (2, 3), xytext=(4, 3), arrowprops={"width": 0.8, "shrink" : 0.1})
ax.text(2, 2, "A Simple text", c="r", fontsize=15)
plt.show()
```

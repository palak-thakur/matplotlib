```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names= np.array(["Megumi", "Sasha", "Sakura", "Inosuke"])
fight= np.array([90, 83, 23, 89])
fun= np.array([63, 99, 56, 99])
```

```python
#line and marker plot
fig, ax= plt.subplots()
ax.plot(names, fight, ls="--", lw=3, marker="P", mfc="r", ms=14, c='g')
ax.set_xlabel("Names of students")
ax.set_ylabel("fighting skills")
ax.set_title("Students Vs fighting skills")
ax.grid(True)
plt.show()
```

```python
#line and marker plot, marker only
fig1, ax1= plt.subplots()
ax1.plot(names, fun, ls="None", marker="p", mfc="b", ms=14)
ax1.set_xlabel("names of students")
ax1.set_ylabel("Fun loving")
ax1.set_title("fun loveing scale")
ax1.grid(True)
plt.show()
```

```python
#bar plot
fig2, ax2= plt.subplots()
cl=["green", "purple", "brown", "c"]
ax2.bar(x= names, height= fight, width= 0.4, color=cl)
ax2.set_xlabel("Names")
ax2.set_ylabel("fighting skills")
ax2.set_title("name v/s fighting skills")
ax2.grid(True)
plt.show()
```

```python
#horizontal bar plot
fig3, ax3= plt.subplots()
cl=["green", "red", "blue", "purple"]
ax3.barh(y=names, width= fight, height=0.4, color= cl)
ax3.set_ylabel("Names")
ax3.set_xlabel("fighting skills")
ax3.set_title("fighting skills v/s names")
ax3.grid(True)
plt.show()
```

```python
#scatter plot
fig4, ax4= plt.subplots()
ax4.scatter([2, 3, 4, 5], [6, 7, 8, 9], s=20, c='black')
plt.show()
```

```python
#pie chart
fig5, ax5= plt.subplots()
cl=["green", "c", "blue", "purple"]
ax5.pie(fight,
       labels= names,
       explode= [0, 0, 0.2, 0],
       autopct= lambda pc: round(pc*np.sum(fight)/100),
       shadow= True,
       colors= cl,
       startangle= 90)
ax5.legend()
ax5.axis('equal')
plt.show()
```

```python
#box plot
fig6, ax6= plt.subplots()
ax6.boxplot([1, 2, 3, 88])
plt.show()
```

```python
#stacked bar chart
fig7, ax7= plt.subplots()
ax7.bar(x= names, height= fight, width=0.4, label="fight")
ax7.bar(x=names, height=fun, width=0.4, label= "fun", bottom= fight)
ax7.set_xlabel("Names")
ax7.set_ylabel("skills")
ax7.set_title("name v/s skills")
ax7.legend()
ax7.grid(True)
plt.show()
```

```python
#grouped bar chart
fig8, ax8= plt.subplots()
ax8.bar(x= names, height= fight, width=-0.4, label="fight", align='edge')
ax8.bar(x=names, height=fun, width=0.4, label= "fun", align='edge')
ax8.set_xlabel("Names")
ax8.set_ylabel("skills")
ax8.set_title("name v/s skills")
ax8.legend()
ax8.grid(True)
plt.show()
```

```python
#sine plot
x= np.linspace(-np.pi, np.pi, 200)
fig9, ax9= plt.subplots()
ax9.plot(x, np.sin(x))
ax9.set_xlabel("angle")
ax9.set_ylabel("sin(x)")
ax9.set_title("sine curve")
ax9.axhline()
ax9.axvline()
ax9.grid(True)
plt.show()
```

```python
#subplots
# 2 sunbplot
fig10, ax10= plt.subplots(nrows=2, ncols=1)
ax10[0].plot(names, fight, ls="--", marker="P", color="blue", mfc='yellow')
ax10[1].plot(names, fun, ls="-", marker="P", color="red", mfc="green")
plt.show()
```

```python
# 4 subplots
fig11, ax11= plt.subplots(nrows=2, ncols=2)
cl=["green", "magenta", "blue", "purple"]
ax11[0,0].plot(names, fight, ls="--", marker="P", color="blue", mfc='yellow')
ax11[0,1].plot(names, fun, ls="-", marker="P", color="red", mfc="green")
ax11[1,0].bar(x= names, height= fight, width= 0.4, color= cl)
ax11[1,1].bar(x=names, height= fun, width=0.4, color= cl)
plt.show()
```

```python
#annotation and text
fig12, ax12= plt.subplots()
ax12.scatter(names, fight, c= "purple")
ax12.annotate("least skilled student",("Sakura", 23), xytext=(2,60), arrowprops={"width": 0.8, "shrink" : 0.1})
plt.show()
```

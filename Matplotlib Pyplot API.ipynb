```python
import numpy as np
import matplotlib.pyplot as plt
```

# Pyplot API

## Bar plot

```python
names = np.array(["Amit", "Raman", "Geeta", "Neeta", "Jay"])
en_marks = np.array([73, 60, 67, 65, 75])  # marks in English
sc_marks = np.array([65, 70, 75, 63, 60])  # Marks in Science
mt_marks = np.array([60, 75, 62, 63, 70])  # Marks in Maths
```

```python
# Bar plot with names in X axis and en_marks on y axis
clr = ['r', "g", 'y', 'c', 'm']
barins = plt.bar(x=names, height=en_marks, color=clr, width=0.4)
plt.grid(visible=True)
plt.xlabel("Names")
plt.ylabel("English marks")
plt.title("Names Vs Marks")
plt.show()
```

```python
# Adjusting alignment
plt.bar(x=names, height=en_marks, color="g", width=-0.4, align="edge")
plt.grid(visible=True)
plt.show()
```

```python
# stacked Bar plot with names in X axis and marks  on y axis
plt.bar(x=names, height=en_marks, color="g", width=0.4, label="English")
plt.bar(x=names, height=sc_marks, color="m", width=0.4, bottom=en_marks, label="Science")
plt.legend()
plt.show()
```

```python
# Grouped bar chart 
xmark = np.arange(names.size)
wd = 1/4
plt.figure(figsize=(10, 5))
plt.bar(x=xmark-wd, height=en_marks, color="g", width=wd, label="English")
plt.bar(x=xmark, height=sc_marks, color="m", width=wd, label="Science")
plt.bar(x=xmark+wd, height=mt_marks, color="b", width=wd, label="Maths")
plt.xticks(xmark, names, rotation=30)
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.title("Names Vs Marks plot")
plt.grid(visible=True)
plt.legend()
plt.show()
```

# Line and Marker plot

```python
# Plot of Names Vs Marks in English
plt.plot(names,
         en_marks,
         color="r",
         ls="-.",
         lw=2,
         marker="p",
         ms=8,
         mfc="g",
         label="English")
plt.show()
```

```python
# With marker & line Names Vs Marks in English & Science
plt.plot(names,
         en_marks,
         color="r",
         ls="-.",
         lw=2,
         marker="p",
         ms=8,
         mfc="g",
         label="English")

plt.plot(names,
         sc_marks,
         color="g",
         ls="--",
         lw=2,
         marker="o",
         ms=8,
         mfc="g",
         label="Science")

plt.legend()
plt.show()
```

```python
# Only Markers Plot of Names Vs Marks in English & Science
plt.plot(names,
         en_marks,
         ls="None",
         marker="p",
         ms=8,
         mfc="g",
         label="English")

plt.plot(names,
         sc_marks,
         ls="None",
         marker="o",
         ms=8,
         mfc="g",
         label="Science")
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.title("Names Vs Marks plot")
plt.legend()
plt.show()
```

# Scatter plot

```python
# Names Vs Marks in English
plt.scatter(names, en_marks, s=30, marker="o", c="r")
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.title("Names Vs Marks plot")
plt.show()
```

```python
# names Vs Marks in English & Science
plt.scatter(names, en_marks, s=30, marker="o", c="r", label="English")
plt.scatter(names, sc_marks, s=30, marker="p", c="g", label="Science")
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.title("Names Vs Marks plot")
plt.legend()
plt.show()
```

# Histogram 

```python
arr = np.array([2, 5, 7, 8, 9, 11, 12, 13, 15, 16, 17, 20])
arr
```

```python
plt.hist(arr, bins=5, color="m")   #by default bins=10, by default mid allignment
plt.xlabel("Values of array")
plt.ylabel("Frequency")
plt.title("Histogram for the array arr")
plt.show()
```

```python
bn = [2, 8, 15, 20]
plt.hist(arr, bins=bn, color="g")   #by default bins=10, by default mid allignment
plt.xlabel("Values of array")
plt.ylabel("Frequency")
plt.title("Histogram for the array arr")
plt.show()
```

# Pie Chart 

```python
# f'{round((pc * np.sum(en_marks))/100)
# Names Vs Marks in English
plt.figure(figsize=(7, 7))
plt.pie(en_marks,                       # data must be 1D array like
        explode=[0, 0, 0.1, 0, 0],
        labels=names,                  # labels should be list
        colors=["r", "g", "b", "m", "c"],
        autopct=lambda pc: f'{round((pc * np.sum(en_marks))/100)}')
plt.legend()
plt.show()
```

# Box Plot

```python
# Box plot of 2 numpy arrays in l
l1 = np.array([2, 4, 8, 3, 5, 10, 24])
l2 = np.array([3, 9, 15, 21, 27, 100, 25])
l3 = [l1, l2]
plt.boxplot(l3)
plt.show()
```

# Annotations & text 

```python
plt.scatter(l1, l2, s=30, marker="o", c="r")
plt.xlabel("Content of array L1")
plt.ylabel("Content of array L2")
plt.title("Scatter plot l1 Vs L2")
plt.annotate("Farthest point", (10.0, 100.0), xytext=(15.0, 100.0), 
             arrowprops=dict(arrowstyle="->"))
plt.text(10, 18, "Scatter plot with annotation", fontsize=15)
plt.show()
```

# Bar plot with names in X axis and en_marks on y axis - with bar labels

```python
barins = plt.bar(x=names, height=en_marks, color="g", width=0.4)
plt.grid(visible=True)
plt.bar_label(barins)
plt.show()
```

# Sine curve with vertical and horizontal line 

```python
x = np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axhline()
plt.axvline()
plt.show()
```

# Multiple plots in one

```python
x = np.linspace(-np.pi, np.pi, 200)
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x), ls="--")
plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x), ls="--")
plt.show()
```
